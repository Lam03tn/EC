```python
import xarray as xr

# Local file path (adjust if needed)
file_path = "C:\\Users\\LamNH\\Downloads\\MERRA2_400.inst1_2d_asm_Nx.20230101.nc4"

import xarray as xr

# Open the dataset
ds = xr.open_dataset(file_path)

# Print dataset information
print(ds)

# Access specific variables (e.g., 2-meter air temperature)
t2m = ds['T2M']  # Replace 'T2M' with the variable of interest
time = ds['time']
lat = ds['lat']
lon = ds['lon']

# Print variable metadata
print(f"T2M: {t2m}")
print(f"Time: {time.values}")
print(f"Latitude: {lat.values}")
print(f"Longitude: {lon.values}")

# Plot the first time step
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.contourf(lon, lat, t2m[0, :, :], cmap="coolwarm")
plt.colorbar(label="Temperature (K)")
plt.title("2-Meter Air Temperature (First Time Step)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

```
